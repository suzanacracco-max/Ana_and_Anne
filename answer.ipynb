```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "category": ["cat", "dog", "mouse", "bird"],
    "value": [10, 0, 0, 0]
})
```

```python
plt.figure(figsize=(6, 4))
data.plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Value")
plt.xticks(ticks=range(len(data)),labels=["cat", "dog", "cat", "bird"])
plt.tight_layout()
plt.show()
```
